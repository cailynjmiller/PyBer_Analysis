```python
%matplotlib inline
```

```python
# Import dependencies
import matplotlib.pyplot as plt
```

```python
# Set the x-axis as months
x_axis = ["Jan","Feb","Mar","Apr","May","Jun","Jul","Aug","Sep","Oct","Nov","Dec"]

# Set the y-axis as total fares in USD accumulated each month
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Create the plot
plt.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt() the other way
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt() ANOTHER way
ax = plt.axes()
ax.plot(x_axis,y_axis)
```

```python
# Create the plot
plt.plot(x_axis,y_axis)
plt.show()
```

```python
# Create the plot and add a label for the legend.
plt.plot(x_axis, y_axis, marker="*", color="blue", linewidth=2, label='Boston')

# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare ($)")

# Set the y limit between 0 and 45.
plt.ylim(0, 45)

# Create a title.
plt.title("PyBer Fare by Month")

# Add the legend.
plt.legend()
```

```python
# Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis)

# Change the line color to green. Change the marker to a diamond. Change the line width to 2 points.
ax.plot(x_axis, y_axis, marker='d', color='green', linewidth=2, label='Boston')

# Add a legend for the city of Boston.
ax.legend()

# Add a title and axes labels.
ax.set_title("PyBer Fair by Month")

# Add a y-axis limit.
ax.set_ylim(0,45)

# Add grid lines.
ax.grid()
```

```python
# Create the plot
plt.bar(x_axis, y_axis, color="green", label='Boston')

# Create labels for the x and y axes
plt.xlabel("Date")
plt.ylabel("Fare($)")

# Create a title
plt.title("PyBer Fair by Month")

# Add the legent
plt.legend()
```

```python
# Create the plot. Change the bar colors to magenta.
plt.barh(x_axis, y_axis, color="magenta", label='Boston')

#Add a legend for the city of Boston.
plt.legend()

#Add a title and axes labels.
plt.title("PyBer Fair by Month")
plt.ylabel("Date")
plt.xlabel("Fare($)")

#Invert the y-axis so that January is at the top.
plt.gca().invert_yaxis()
```

```python
# Create the plot with ax.plt()
fig, ax=plt.subplots()
ax.bar(x_axis,y_axis)
```

```python
# Create the plot with ax.plt(). Change the bar color to cyan. Switch the axis so that the Fare($) data is on the x-axis.
fig, ax=plt.subplots()
ax.barh(x_axis,y_axis, color="cyan", label="Chicago")

# Add a legend for the city of Chicago.
ax.legend()

# Add a title and axes labels.
ax.set_title("PyBer Fares by Month")
ax.set_xlabel("Fare ($)")
ax.set_ylabel("Month")

# Invert the y-axis data.
ax.invert_yaxis()
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
plt.plot(x_axis, y_axis, "o")
```

```python
# Change the color of the markers to red. Switch the axis so that the Fare($) data is on the x-axis.
plt.scatter(y_axis, x_axis, color="red", label='Chicago')

#Add a legend for the city of Chicago.
plt.legend()

#Add a title and axes labels.
plt.ylabel('Month')
plt.xlabel('Fares ($)')
plt.title('PyBer Fares by Month')

#Add a limit for the x-axis data.
plt.xlim(0,45)

#Invert the y-axis so that January is at the top.
plt.gca().invert_yaxis()
```

```python
plt.scatter(x_axis, y_axis, s=y_axis)
```

```python
y_axis_larger = []
for data in y_axis:
    y_axis_larger.append(data*3)
```

```python
plt.scatter(x_axis, y_axis, s=y_axis_larger)
```

```python
plt.scatter(x_axis, y_axis, s = [i*3 for i in y_axis])
```

```python
fig, ax=plt.subplots()

# Change the color of the markers to sky blue. Change the size of the markers to 5 times each data point. 
#Make the color 20% transparent.
# Add a black edge color to the circles.
# Make the linewidth of the circles 2 points.
# Switch the axis so that the Fare($) data is on the x-axis.
ax.scatter(y_axis,x_axis, s=[i*5 for i in y_axis], edgecolor='black', color="blue", label='Boston', alpha=.2, linewidth=2)

# Add a legend for the city of Boston.
ax.legend()

# Add a title.
ax.set_title('Pyber Fares by Month')

# Add a limit for the x-axis data.
ax.set_xlim(0,45)

# Inver y axis so jan is on top
ax.invert_yaxis()
```

```python
plt.pie(y_axis, labels=x_axis)
plt.show()
```

```python
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.subplots(figsize=(8, 8))
plt.pie(y_axis,
    explode=explode_values,
    colors=colors,
    labels=x_axis,
    autopct='%.1f%%')

plt.show()
```

```python
# Add a percentage to one decimal place to each wedge of the pie.
# Increase the figure size to 8x8.
# Explode the two highest percentage months.
# Add a shadow.
# Add a start angle so that January is at the top.
# Reverse the order so that the month order is in a clockwise direction.
# Add new colors of your own choosing or use the colors from the previous pie chart.

colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
explode_values = (0,0,0.2,0,0,0,0.2,0,0,0,0,0)
fig, ax = plt.subplots(figsize=(8,8))
ax.pie(y_axis,labels=x_axis, autopct='%.1f%%', explode=explode_values, shadow=True, startangle=100, counterclock=False, colors=colors)
plt.show()
```
